## SELECT * FROM players WHERE salary > 1000000

## SELECT * FROM players WHERE position IN ('C','LW','RW')

```python
players[players.position.isin(['C','LW','RW'])]
```

```python
players.loc[players.position.isin(['C','LW','RW'])]
```

## SELECT * FROM players WHERE position NOT IN ('C','LW','RW')

```python
players[~players.position.isin(['C','LW','RW'])]
```

```python
players.loc[~players.position.isin(['C','LW','RW'])]
```

## SELECT * FROM players WHERE team = 'VAN' OR position = 'C'

```python
players[(players.team=='VAN') | (players.position=='C')]
```

```python
players.query('(team=="VAN") or (position == "C")')
```

## SELECT * FROM players WHERE team = 'VAN' AND position = 'C'

```python
players[(players.team=='VAN') & (players.position=='C')]
```

```python
players.query('(team=="VAN") and (position == "C")')
```

## Pro trick: Return all columns that contain a given value i.e. tell me which columns contain a value in my list

```python
players.loc[:,players.isin(['VAN','MTL']).any()]
```

## SELECT * FROM players WHERE first_name LIKE 'Da%'
